```
path = 'jaffe/jaffe'
```

```
from fastai import *
from fastai.vision import *
```

```
!pip install "torch==1.4" "torchvision==0.5.0"
```

```
np.random.seed(42)
tfms = tfms = get_transforms(flip_vert=False, do_flip=True, max_rotate = 10.0, max_zoom = 1.1 )
data = ImageDataBunch.from_folder(path=path, ds_tfms = tfms, train=".",valid_pct = 0.20,
         size=256, bs = 35 , num_workers=4).normalize(imagenet_stats)
```

```
data.classes, data.c, len(data.train_ds), len(data.valid_ds)
```

```
data.show_batch(rows=4,figsize=(10,10))
```

```
learner.lr_find()
learner.recorder.plot()
```

```
learner=cnn_learner(data,models.vgg16_bn,metrics=accuracy)
learner.fit_one_cycle(30)
```

```
learner.save('stage1')
```

```
learner.lr_find()
learner.recorder.plot()
```

```
learner.unfreeze()
```

```
learner.fit_one_cycle(30,max_lr=slice(1e-3,1e-2))
```

```
learner.save('stage2')
```

```
learner.lr_find()
learner.recorder.plot()
```

```
interp = ClassificationInterpretation.from_learner(learner)
interp.plot_top_losses(9, figsize=(15,11))
```

```
interp.plot_confusion_matrix(figsize=(12,12), dpi=60)
```

```
learner.save('stage2')
learner.unfreeze()
learner.fit_one_cycle(5, max_lr=slice(1e-4,1e-3))
```

```
learner.save('stage3')
```

```
learner.freeze()
```
